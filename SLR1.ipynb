```python


df = pd.read_csv('/content/drive/MyDrive/Machine Learning GrowTech/ml dataset/Salary Data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df.isnull().sum()
```

```python
#remove uunnecessary column
df.drop(columns = ['Age','Gender','Education Level','Job Title'], axis =1,inplace=True)
df
```

```python
#Feature Selection
x =df[['Years of Experience']]
y =df['Salary']
```

```python
x
```

```python
y
```

```python
# Normalize the data
from sklearn.preprocessing import StandardScaler
st_scale = StandardScaler()
x_scaled = st_scale.fit_transform(x)
```

```python
 # train the split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(  x_scaled,y ,test_size=0.3)
```

```python
print(x_train.shape,x_test.shape),
print(y_train.shape,y_test.shape)
```

```python
# Model Building
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Prediction
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
